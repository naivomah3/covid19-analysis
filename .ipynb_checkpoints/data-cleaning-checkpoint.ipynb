```python
import pandas as pd 
import numpy as np

import warnings
warnings.simplefilter('ignore')

# Full display rows option
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
%matplotlib inline
```

```python
confirmed_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
death_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
recovered_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'
```

```python
confirmed = pd.read_csv(confirmed_url) #.fillna("N/A", inplace = True)
deaths = pd.read_csv(death_url)
recovered = pd.read_csv(recovered_url)
```

```python
confirmed['Province/State'].fillna('N/A', inplace=True)
deaths['Province/State'].fillna('N/A', inplace=True)
recovered['Province/State'].fillna('N/A', inplace=True)
```

```python
# Checking inconsistency as the Recovered dataset doesn't have the same shape as 
# Death and Confirmed datasets
def dataframe_difference(df1, df2, which=None):
    """Find rows which are different between two DataFrames."""
    comparison_df = df1.merge(df2,
                              indicator=True,
                              how='outer')
    if which is None:
        diff_df = comparison_df[comparison_df['_merge'] != 'both']
    else:
        diff_df = comparison_df[comparison_df['_merge'] == which]
    return diff_df
```

```python
cols_to_confirm = ['Province/State', 'Country/Region'] # used as index 
inner_confirm_rec = dataframe_difference(confirmed[cols_to_confirm],
                                         recovered[cols_to_confirm],
                                         which='both')
inner_death_rec = dataframe_difference(deaths[cols_to_confirm],
                                       recovered[cols_to_confirm],
                                       which='both')
```

```python
# Fetch and Reshape
def melt_data(df, case_type):
    df_melt = df.melt(id_vars=['Province/State', 
                                 'Country/Region', 
                                 'Lat', 
                                 'Long'])
    
    df_melt.rename(columns={'variable': 'Date', 
                              'value': case_type,
                              'Province/State': 'Province_State',
                              'Country/Region': 'Country_Region'
                             }, 
                     inplace=True)
    
    return df_melt
    
```

```python
df_confirmed = melt_data(confirmed, "Confirmed")
df_recovered = melt_data(recovered, "Recovered")
df_deaths = melt_data(deaths, "Death")
```

```python
# Confirm datasets have the same shape 
df_confirmed
df_deaths
df_recovered
```

```python
# Join Confirmed & Deaths 
df = df_confirmed.join(df_deaths['Death'])
```

```python
df.head()
```

```python
new_df = pd.merge(df_recovered, df)
```

```python
new_df
```

```python
# Fiigure out the percentage of missing values 
total = new_df.isnull().sum().sort_values(ascending=False) 
percentage = (new_df.isnull().sum() / new_df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percentage], axis=1, keys=['Total', 'Percentage'])
missing_data[missing_data['Total'] > 0] 
```
